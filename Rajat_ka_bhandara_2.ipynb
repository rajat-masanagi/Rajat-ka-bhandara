### Neural Network Approach

#### Importing Libraries

```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.metrics import f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

#### Reading the data

```python
train = pd.read_csv('train 2.csv')
test = pd.read_csv('test 2.csv')
```

##### Preprocessing

```python
features = ['DRUGTYPE', 'Drug_high_status', 'Drug_Status',
            'Disease_of_highest_status', 'TARGNAME', 'GENENAME',
            'BIOCLASS', 'Disease'] 

X_train_sample = train[features]
y_train_sample = train['Target_Status']

X_test = test[features]
X_train_sample = X_train_sample.astype(str)
X_test = X_test.astype(str)
ohe = OneHotEncoder()
X_encoded_combined = ohe.fit_transform(pd.concat([X_train_sample, X_test], keys=['train', 'test']))

X_encoded_train = X_encoded_combined[:len(X_train_sample)]

le = LabelEncoder()
y_train_sample = le.fit_transform(y_train_sample)

y_encoded_train = to_categorical(y_train_sample)  
```

##### Model Building

```python

model = Sequential()
model.add(Dense(512, input_dim=X_encoded_train.shape[1], activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(y_encoded_train.shape[1], activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
initial_epochs = 70
# early_stopping = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)
```

```python

history = model.fit(X_encoded_train, y_encoded_train,epochs=initial_epochs, batch_size=32)
```

##### Prediction

```python
test_predictions = model.predict(X_encoded_combined[len(X_train_sample):])
test_predictions_classes = np.argmax(test_predictions, axis=1)
```

```python
ypred=le.inverse_transform(test_predictions_classes)
ypred
```

```python
predictions_df = pd.DataFrame({
    'ID': test['ID'],
    'Prediction': ypred
})

predictions_df.to_csv('submission_Final_nn.csv', index=False)
print("Predictions saved to submission_Final_nn.csv")
```
